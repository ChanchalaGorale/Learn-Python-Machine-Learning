# Data Science Plots with Seaborn & Matplotlib


```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set(style="whitegrid")
```

```python

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")
flights = sns.load_dataset("flights")
```

## 1. Histogram - Distribution of a Single Variable

```python

sns.histplot(data=tips, x="total_bill", kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.show()
```

## 2. Box Plot - Comparing Distributions Across Categories

```python

sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Boxplot of Total Bill by Day and Gender")
plt.show()
```

## 3. Violin Plot - Distribution + Density

```python

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Violin Plot of Total Bill by Day and Gender")
plt.show()
```

## 4. Bar Plot - Mean Comparison

```python

sns.barplot(data=tips, x="day", y="tip", estimator=np.mean, ci="sd")
plt.title("Average Tip by Day")
plt.show()
```

## 5. Count Plot - Frequency of Categories

```python

sns.countplot(data=tips, x="day", hue="sex")
plt.title("Count of Records by Day and Gender")
plt.show()
```

## 6. Scatter Plot - Relationship Between Two Variables

```python

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Total Bill vs Tip by Time")
plt.show()
```

## 7. Pair Plot - Multiple Relationships at Once

```python

sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

## 8. Correlation Heatmap - Matrix of Relationships

```python
penguins.info()
```

```python
corr = penguins.select_dtypes(include='number').dropna().corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix - Penguins Dataset")
plt.show()
```

## 9. Line Plot - Trends Over Time

```python
plt.figure(figsize=(20,20))
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Monthly Passengers Heatmap (1949-1960)")
plt.show()
```

## 10. Joint Plot - Bivariate with Marginals

```python

sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
plt.suptitle("Joint Plot: Total Bill vs Tip", y=1.02)
plt.show()
```
